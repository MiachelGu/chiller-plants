```r
get_df <- function(path, pattern) {
  # read data
  files <- dir(path=path, pattern=pattern, full.names=TRUE);
  df <- do.call(rbind, lapply(files, read.csv));
  
  # update col names
  cols <- colnames(df);
  cols[1] <- "timestamp";
  colnames(df) <- cols;
  
  # update data types
  df[df == "\\N"] <- NA;
  
  df
}

clc <- function() {
  cat("\014");
}

df <- get_df("../data/insead", "*21*.csv")
```
